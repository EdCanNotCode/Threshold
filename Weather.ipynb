```python
import json
import os

import requests
from dotenv import load_dotenv
import matplotlib.pyplot as plt

load_dotenv()

GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
assert GOOGLE_API_KEY is not None, "GOOGLE_API_KEY is not set"

LOCATION = {
    "latitude": 51.8642,
    "longitude": -2.2382,
}
```

```python
resp = requests.get(
    "https://weather.googleapis.com/v1/forecast/days:lookup",
    params={
        "key": GOOGLE_API_KEY,
        "location.latitude": LOCATION["latitude"],
        "location.longitude": LOCATION["longitude"],
        "days": 1,
    }
)

weather_forecast = resp.json()["forecastDays"][0]
print(json.dumps(weather_forecast, indent=2))
```

```python
resp = requests.get(
    "https://pollen.googleapis.com/v1/forecast:lookup",
    params={
        "key": GOOGLE_API_KEY,
        "location.latitude": LOCATION["latitude"],
        "location.longitude": LOCATION["longitude"],
        "plantsDescription": False,
        "days": 5,
    }
)

pollen_forecast = resp.json()["dailyInfo"][1]
print(json.dumps(pollen_forecast, indent=2))
```
